```python
labels = ['frogs','dogs','cats', 'birds']
sizes = [15,30,45,10]
explode = (0,0.1,0,0) #the explode is how the dogs section is out further
fig1,ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True,startangle=90) 
ax1.axis('equal') #this makes sure it is a circle
```
